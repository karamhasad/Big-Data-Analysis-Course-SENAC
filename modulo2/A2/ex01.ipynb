```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('pesquisa_satisfacao.csv', sep =';')

#display é opcional, basta por a variavel. 

#display(df)
```

```python
media_satisfacao = df.groupby('Satisfação')['Tempo de Espera (min)'].mean().reset_index()
media_satisfacao['Tempo de Espera (min)'] = round(media_satisfacao['Tempo de Espera (min)'] , 2)
media_satisfacao
```

```python
df_anp = pd.read_csv('ANP2025.csv', sep = ';')
df_anp
```

```python
df_anp_rj = df_anp.loc[df_anp['Estado - Sigla'] == 'RJ']
df_anp_rj
```

```python
df_anp_rj['Valor de Venda'] = df_anp_rj['Valor de Venda'].str.replace(',', '.').astype(float)
```

```python
media_munip_rj = df_anp_rj.groupby('Municipio')['Valor de Venda'].mean().reset_index()
media_munip_rj = media_munip_rj.sort_values(by='Valor de Venda', ascending=False)
media_munip_rj
```

```python
array_valor_venda = np.array(media_munip_rj['Valor de Venda'])
```

```python
media = np.mean(array_valor_venda)
mediana = np.median(array_valor_venda)

q1 = np.quantile(array_valor_venda, 0.25)
q2 = np.quantile(array_valor_venda, 0.50)
q3 = np.quantile(array_valor_venda, 0.75)

print(f'A média é {media}')
print(f'A mediana é {mediana}')
print(f'Primeiro quartil (Q1): {q1}')
print(f'Segundo quartil (Q2, Mediana): {q2}')
print(f'Terceiro quartil (Q3): {q3}')
```

```python
munic_acima_q3 = media_munip_rj.loc[media_munip_rj['Valor de Venda'] > q3]
munic_acima_q3
```

```python
df_anp_bairros_cap = df_anp_rj.loc[(df_anp_rj['Municipio'] == 'RIO DE JANEIRO') & (df_anp_rj['Produto'] == 'ETANOL')]
df_anp_bairros_cap
```

```python
media_bairro_rj = df_anp_bairros_cap.groupby('Bairro')['Valor de Venda'].mean().reset_index()
media_bairro_rj = media_bairro_rj.sort_values(by='Valor de Venda', ascending=False)
media_bairro_rj
```

```python
array_valor_etanol_capital = np.array(media_bairro_rj['Valor de Venda'])
```

```python
media_etanol = np.mean(array_valor_etanol_capital)
mediana_etanol = np.median(array_valor_etanol_capital)

q1_etanol = np.quantile(array_valor_etanol_capital, 0.25)
q2_etanol = np.quantile(array_valor_etanol_capital, 0.50)
q3_etanol = np.quantile(array_valor_etanol_capital, 0.75)

print(f'A média é {media_etanol}')
print(f'A mediana é {mediana_etanol}')
print(f'Primeiro quartil (Q1): {q1_etanol}')
print(f'Segundo quartil (Q2, Mediana): {q2_etanol}')
print(f'Terceiro quartil (Q3): {q3_etanol}')
```

```python
bairros_etanol = df_anp_bairros_cap.groupby('Bairro')['Valor de Venda'].mean().reset_index()
bairros_etanol = bairros_etanol.sort_values(by='Valor de Venda', ascending=False)
bairros_etanol
```

```python
bairros_cap_etanol_q3 = media_bairro_rj.loc[media_bairro_rj['Valor de Venda'] > q3_etanol ]
bairros_cap_etanol_q3
```
